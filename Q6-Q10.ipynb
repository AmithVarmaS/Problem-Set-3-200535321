# Question 6

Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

us_mrg_div = pd.read_csv('us-marriages-divorces-1867-2014.csv')
years = us_mrg_div['Year'].values
mrg = us_mrg_div['Marriages_per_1000'].values
div = us_mrg_div['Divorces_per_1000'].values
plt.plot(years,mrg, years,div)
plt.title('Line plot depicting the number of weddings and divorces per capita in the United States between 1867 and 2014')
plt.xlabel('Years')
plt.ylabel('Marriages and Divorces per capita')
plt.legend(["blue", "orange"], loc ="lower right")
plt.show()
```

# Question 7

Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.

Don't forget to label your axes!

```python
import pandas as pd
import numpy as np

us_mrg_div = pd.read_csv('us-marriages-divorces-1867-2014.csv')
us_mrg_div = us_mrg_div[us_mrg_div['Year'].apply(lambda x: x in [1900, 1950, 2000])]

years = us_mrg_div['Year'].values
mrg = us_mrg_div['Marriages_per_1000'].values
div = us_mrg_div['Divorces_per_1000'].values

n = np.arange(len(years))
# Passing the parameters to the bar function
# Using 'n' to align the bars side by side
plt.bar(n, mrg, color = 'black', width = 0.25)
plt.bar(n + 0.25, div , color = 'g', width = 0.25)
plt.legend(['Marriages per capita', 'Divorces per capita'])

plt.xticks([i + 0.25 for i in range(3)], [1900, 1950, 2000])
plt.title("Bar plot representing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000")
plt.xlabel('Years')
plt.ylabel('Count of marraiges/divorces ')

plt.show()

```

# Question 8

Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.

Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt
actor_kills = pd.read_csv('actor_kill_counts.csv')
actor_kills_desc = actor_kills.sort_values('Count',ascending=False)

actors = actor_kills_desc['Actor'].values
kills = actor_kills_desc['Count'].values
print(kills)

plt.barh(actors,kills)
plt.title('Horizontal Bar Plot of Hollywood Actors and their kill count')
plt.ylabel('Actors')
plt.xlabel('Kill Count')
plt.show()
```

# Question 9


Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.

Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

roman_emperors = pd.read_csv('roman-emperor-reigns.csv')
assassinated_emperors = roman_emperors[
    roman_emperors['Cause_of_Death'].apply(lambda x: 'assassinated' in x.lower())]


no_emp_assassinated = len(assassinated_emperors)
other_deaths = len(roman_emperors) - no_emp_assassinated


emperor = assassinated_emperors["Emperor"]
cause = assassinated_emperors["Cause_of_Death"]
plt.pie(range(len(cause)), labels=emperor,autopct='%1.1f%%', startangle=360, radius=0.05 * 40,rotatelabels = 270)
fig = plt.figure(figsize=[2, 2])
```

arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.

Don't forget to label your axes!
